```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('netflix_data.csv')
```

```python
df.head()
```

```python
df['cast'].replace(to_replace=np.nan, value='None', inplace=True)
```

```python
val = df['cast']

def cast_num(val):
    if val== 'None':
        return 0
    else:
        return len(val.split(', '))
```

```python
df['num_of_cast'] = df['cast'].apply(cast_num)
```
